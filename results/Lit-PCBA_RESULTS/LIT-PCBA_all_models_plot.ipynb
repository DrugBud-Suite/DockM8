```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load literature data
litpcbaliterature = pd.read_excel('/home/alacournola/DockM8/results/Lit-PCBA_RESULTS/Lit-PCBA_literature.xlsx')
# Load DockM8 data
litpcbaef1 = pd.read_csv('/home/alacournola/lit-pcba/LIT-PCBA_RESULTS/results_dockm8_EF_1%.csv')
#Determine DockM8-max
max_values = litpcbaef1.iloc[:, 3:].max()
column_names = litpcbaef1.columns[3:]
max_df = pd.DataFrame({'Target': column_names, 'Model' : 'DockM8-max', 'EF1': max_values}).reset_index(drop=True)
#Determine DockM8-best
litpcbaef1['average'] = litpcbaef1.iloc[:, 3:].mean(axis=1)
litpcbaef1 = litpcbaef1.sort_values(by='average', ascending=False)
litpcbaef1 = litpcbaef1.head(1)
litpcbaef1 = litpcbaef1.drop(columns=['average'])
best_df = pd.melt(litpcbaef1, value_vars=litpcbaef1.columns[3:], var_name='Target', value_name='EF1')
best_df['Model'] = 'DockM8-best'
best_df = best_df[['Target', 'Model', 'EF1']]
#concatenate all data
litpcbaliterature_filtered = litpcbaliterature[litpcbaliterature['Target'].isin(column_names)]
litpcbafinal = pd.concat([litpcbaliterature_filtered, best_df, max_df])
litpcbafinal = litpcbafinal.drop(columns=['Source'])
litpcbafinal = litpcbafinal.sort_values(by = 'Target', ascending = True)
ef1_df = pd.melt(litpcbafinal, id_vars=['Model'], value_vars=['EF1'], var_name='EF1', value_name='Score')
unique_models = ef1_df['Model'].unique()
median_by_method_ef1 = litpcbafinal.groupby('Model')['EF1'].median().sort_values(ascending=False).index
print(median_by_method_ef1)
# Assign colors based on the median values
color_palette = sns.color_palette("RdYlGn", n_colors=len(median_by_method_ef1))[::-1]
color_dict = dict(zip(median_by_method_ef1, color_palette))
for key in color_dict.keys():
    if 'DockM8' in key:
        color_dict[key] = 'orange'
ef1_df['color'] = ef1_df['Model'].map(color_dict)
ef1_df.loc[ef1_df['Model'].str.contains('DockM8'), 'color'] = 'orange'
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(6, 10), dpi=600)
plot = sns.boxplot(x='Score', y='Model', hue='Model', data=ef1_df, 
            order = median_by_method_ef1, 
            native_scale = True, palette=color_dict, legend=False,
            linewidth=1,
            fliersize=1, 
            whis=(10, 90), 
            medianprops={"color": "black", "linewidth": 1.5})
#plt.title('Box Plot of EF Values for Each Model on DEKOIS 2.0')
plt.ylabel(None)
plt.xlabel('EF (%)')
plt.xticks(fontsize=10)

for idx, label in enumerate(plot.get_yticklabels()):
    if idx < 2:
        label.set_fontweight('bold')
    else:
        label.set_fontweight('normal')

plt.savefig("/home/alacournola/DockM8/results/plots/litpcba_all_models_EF1.png", transparent=False, bbox_inches='tight')
plt.show()
```
